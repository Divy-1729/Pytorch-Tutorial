[torch.Tensor](https://docs.pytorch.org/docs/stable/tensors.html) describes all the details about this class.

## Manipulating Tensors

```python
# Tensor-scalar operations

sixth += 10
seventh -= 10
eight *= 6
ninth /= 25

sixth, seventh, eight, ninth


```

```python
# Matmul

# For matmul the math has to hold, so the inner dimensions must match, and the resultant matrix with have the outer dimensions.

# so (a,b) X (b,c) will give a (a,c) matrix

# Shapes need to be in the right way  so this will return an error
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)

tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]], dtype=torch.float32)

torch.matmul(tensor_A, tensor_B) # (this will error)
```

```python
# We fix this with a transpose!

# matrix.T gives the transpose of a matrix.

torch.matmul(tensor_A, tensor_B.T)
```

```python
# Aggregation Operations

eleventh = torch.arange(0,50,10)

eleventh
```

```python
print(f"Minimum: {eleventh.min()}")
print(f"Maximum: {eleventh.max()}")
# print(f"Mean: {eleventh.mean()}") # this will error
print(f"Mean: {eleventh.type(torch.float32).mean()}") # won't work without float datatype
print(f"Sum: {eleventh.sum()}")
```

```python
# Returns index of max and min values
print(f"Index where max value occurs: {eleventh.argmax()}")
print(f"Index where min value occurs: {eleventh.argmin()}")
```

```python
# Reshaping a tensor
alpha = torch.arange(1.,8.,1)
alpha, alpha.shape
```

```python
# add an extra dim

beta = alpha.reshape(1,7)
beta
```

```python
# stack tensors

gamma = torch.stack([beta, beta], dim = 0)
delta = torch.stack([beta,beta], dim = 1)
eps = torch.stack([beta, beta], dim = 2)
gamma, delta, eps
```

`gamma = torch.stack([beta, beta], dim = 0) `stacks the tensors along a new dimension at index 0 (the first dimension). Since beta is a 1x7 tensor, stacking two beta tensors with dim=0 results in a 2x1x7 tensor. This means it adds new 'rows' of the beta tensors on top of each other.

`delta = torch.stack([beta,beta], dim = 1) `stacks the tensors along a new dimension at index 1 (the second dimension). For a 1x7 beta tensor, stacking with dim=1 results in a 1x2x7 tensor. This effectively adds new 'columns' of the beta tensors next to each other within the existing first dimension.

`eps = torch.stack([beta,beta], dim = 2) `stacks the tensors along a new dimension at index 2 (the third dimension). For a 1x7 beta tensor, stacking with dim=1 results in a 1x7x2 tensor.

As we can see, the first dimension of 1 is making things a little complicated here. Essentially we have a row that contains just 1 point `(which is just [[1.,2.,3.,4.,5.,6.,7.]]`. We can remove this

```python
# remove dim of one

phi = eps.squeeze()
phi, phi.shape
```

```python
# reverse this process

sigma = phi.unsqueeze(dim = 0)

torch.equal(sigma, eps)
```

## Indexing

How do we select data from Tensors?

```python
theta = phi.reshape(2,7,1)
theta
```

Indexing always goes from outermost bracket to innermost brackets

```python
print("First row : \n",theta[0])
print("-"*60)
print("First row first column", theta[0][0])
print("-"*60)
print("First row first column first element", theta[0][0][0])
```

You can also extract entire dimensions for a particular index

```python
theta[:,0] # all rows, first column
```

```python
theta[:, 0:4,:] #all rows, columns 1 to 4, all elements
```

```python
theta[:,:,0] # all rows, all columns, first (and only) element :remember size is (2,7,1)
```
